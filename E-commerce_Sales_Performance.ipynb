```python
df.drop(columns=['Unnamed: 22', 'index'], inplace=True, errors='ignore')
```

```python
df.info()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%m-%d-%y', errors='coerce')
```

```python
df['ship_postal_code'] = df["ship_postal_code"].astype('Float64').astype('string')
```

```python
df.isnull().sum()
```

```python
df['courier_status'] = df['courier_status'].fillna("Unknown")
```

```python
df["currency"] = df["currency"].fillna("Unknown")
```

```python
df['amount'] = df['amount'].fillna(df['amount'].mean())
```

```python
df['promotion_ids'] = df['promotion_ids'].fillna("unknown")
```

```python
df['ship_postal_code'] = df['ship_postal_code'].fillna("Unknown")
```

```python
df[['ship_city','ship_state','ship_country']]= df[['ship_city','ship_state','ship_country']].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.to_csv("Cleaned_Amazon_Dataset.csv")
```
